```python
import numpy as np
import requests
import io
import PIL
import matplotlib.pyplot as plt

#    numpy array 
url = 'https://cdn.pixabay.com/photo/2016/10/17/07/53/busan-night-scene-1747130_1280.jpg'
img = np.array(PIL.Image.open(io.BytesIO(requests.get(url).content)))/255

# Red, Green, Blue  
R = img[:,:,0]
G = img[:,:,1]
B = img[:,:,2]

# Gray  
Gray = 0.2989 * R + 0.5870 * G + 0.1140 * B

# Gray  
plt.imshow(Gray, cmap='gray')
plt.axis('off')
plt.show()
```

```python

ch_min = np.min(R[0:587, 948:1210])
ch_max = np.max(R[0:587, 948:1210])

result = np.where((ch_min < R) & (R < ch_max), Gray, R)

Ch_min = np.min(G[0:587, 948:1210])
Ch_max = np.max(G[0:587, 948:1210])

Result = np.where((Ch_min < G) & (G < Ch_max), Gray, G)

cH_min = np.min(B[0:587, 948:1210])
cH_max = np.max(B[0:587, 948:1210])

resulT = np.where((cH_min < B) & (B < cH_max), Gray, B)

#  R, G, B  
modified_img = np.stack((result, Result, resulT), axis=-1)

#  
plt.imshow(modified_img)
plt.axis('off')
plt.show()

```
